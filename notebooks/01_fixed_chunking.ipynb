# CHARACTER BASED CHUNKING

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

document = """
The Rise of AI-Powered Travel Assistants
Overview
AI-powered travel assistants are transforming how people plan, book, and experience travel.Instead of manually searching across dozens of websites, travelers can now rely on conversational 
AI systems to recommend destinations, compare prices, suggest itineraries, and provide real-time updates.Trip Planning
Modern travel assistants analyze user preferences such as budget, travel history, preferred climate,food interests, and trip duration to generate personalized recommendations. 
For example, a traveler interested in beaches, local food, and short weekend trips may receive recommendations for coastal cities with direct flight options and affordable accommodations.\n
Booking and Reservations: AI systems can integrate with airline APIs, hotel providers, and transportation services to automate 
the booking process. Users can ask questions in natural language such as:“Find me the cheapest flights to Singapore next weekend”
or “Book a business hotel near the conference center.”
Real-Time Assistance:During travel, AI assistants provide updates related to flight delays, gate changes, weather alerts, 
currency conversion, and local transportation availability.Some systems also provide multilingual translation support for international travelers.
Challenges
Despite the convenience, AI travel assistants face several challenges:
- Inaccurate recommendations caused by outdated data
- Limited understanding of complex travel preferences
- Privacy concerns around location tracking and booking history
- Difficulty handling unexpected disruptions during multi-city travel

Future Possibilities
The future of AI in travel may include:
- Fully voice-controlled travel planning
- AI-generated dynamic itineraries based on live events
- Augmented reality city navigation
- Emotion-aware recommendation systems
- Personalized travel companions powered by large language models

Conclusion
As AI systems continue improving, travel planning may become increasingly conversational, proactive, 
and personalized. However, maintaining trust, transparency, and accurate recommendations will remain 
critical for widespread adoption.
"""
```

```python
# small chunk size without overlaps
splitter_without_overlap = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=0
)
chunks_without_overlap  = splitter_without_overlap.split_text(document)
```

```python
# small chunk size with overlaps
splitter_with_overlap = RecursiveCharacterTextSplitter(
    chunk_size=250,
    chunk_overlap=100
)
chunks_with_overlap  = splitter_with_overlap.split_text(document)
```

### COMPARISON BETWEEN WIHTOUT AND WITH OVERLAP

```python
for i in range(3):
    print(f"\nWITHOUT OVERLAP - Chunk {i+1}")
    print(chunks_without_overlap[i])

    print(f"\nWITH OVERLAP - Chunk {i+1}")
    print(chunks_with_overlap[i])

    print("-" * 50)
```

## observations

### Without overlap, important context gets split between chunks, making retrieval less reliable.

### With overlap, chunks share a small amount of repeated context, helping preserve semantic continuity between sections.
